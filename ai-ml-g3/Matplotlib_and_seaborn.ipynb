```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x= [1,2,3,4,5]
y = [10,20,30,40,50]
# figsize
plt.figure(figsize=(3,4))
plt.plot(x,y,color="c",linestyle="-",linewidth=3,marker="o",markerfacecolor="green" ,alpha=0.8)
# x and y labels
plt.xlabel("x axis")
plt.ylabel("y axis")
# title
plt.title("basic line plot ")
plt.show()
```

```python
print(plt.style.available)
```

```python
plt.style.use("dark_background")
x= [1,2,3,4,5]
y = [10,20,30,40,50]
plt.figure(figsize=(3,4))
plt.plot(x,y,color="c",linestyle="-",linewidth=3,marker="o",markerfacecolor="green",alpha=0.9)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("line plot")
plt.show()
```

```python
# bar plot - plt.bar()
x= ["A","B","C","D"]     # sort.values()
y = [10,20,30,40]
plt.figure(figsize=(3,4))
plt.barh(x,y,color='c',edgecolor="white",linewidth=2)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("bar plot")
plt.show()
```

```python
# scatter plot
import numpy as np
x= np.random.randint(1,50,20)
y = np.random.randint(1,50,20)
plt.figure(figsize=(3,4))
plt.scatter(x,y,alpha= 0.9)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("scatter plot")
plt.show()
```

```python
# plt.hist (histogram)
x= np.random.rand(50)
plt.figure(figsize=(3,4),edgecolor='black')
plt.hist(x,bins=20)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("histogram plot")
plt.show()
```

```python
x= [1,2,3,4,5]
y=[10,20,30,40,50]
z=[5,10,15,20,25]
# first subplot
plt.subplot(1,2,1)
plt.plot(x,y,color="yellow")

# second subplot
plt.subplot(1,2,1)
plt.plot(x,z,color="red")

plt.show()
```

```python
# heatmap
x = np.random.rand(4,4)
plt.figure(figsize=(3,4))
plt.imshow(x,cmap="coolwarm",interpolation= "bicubic")
plt.colorbar()
plt.title("heatmap title")
plt.show()
```

```python
df['Date']= pd.to_datetime(df['Date'])
plt.figure(figsize=(10,5))
plt.plot(df["Date"],df["Total Revenue"],marker ='o')
plt.xlabel("date")
plt.ylabel("revenue")
plt.title("revenue generated over time")
plt.show()
```

```python
s= df.groupby("Product Category")["Total Revenue"].sum().sort_values()
plt.figure(figsize=(10,5))
plt.bar(s.index,s.values)
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.show()
```

```python
plt.hist(df['Total Revenue'],bins=30,alpha=0.8,edgecolor= "black")
plt.xlabel("total Revenue")
plt.ylabel("Frequency")
plt.title("revenue")
plt.show()
```

```python
# pie  chart
region_revenue = df.groupby("Region")["Total Revenue"].sum()
plt.pie(region_revenue,labels= region_revenue.index,autopct="%1.1f%%", colors= ['b',"c","orange"])
plt.title("region revenue")
plt.show()
```

```python
# seaborn library
import seaborn as sns
```

```python
sns.histplot(df['Total Revenue'],bins=20)
plt.title("revenue")
plt.xlabel("revenue total")
plt.ylabel("revenue count")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr,cmap="coolwarm")
```

```python
# box plot
plt.figure(figsize=(10,6))
sns.boxplot(x="Product Category", y="Total Revenue",data = df)
```

```python
from numpy import mean
sns.barplot(x="Region",y="Total Revenue",data =df,estimator=mean)
```

```python
count_ppl = df.groupby('Payment Method')['Transaction ID'].count()

sns.barplot(count_ppl)
```

```python
sns.countplot(x="Payment Method",data = df)
```

```python
sns.countplot(x="Product Category",data =df)
```

```python

sns.barplot(x='Total Revenue', y='Product Name', data=df.head(10).sort_values(by='Total Revenue', ascending=False))
```

```python
# Top 10 Products based on revenue
top10 = df.groupby("Product Name")["Total Revenue"].sum().nlargest(10).reset_index()
sns.barplot(data = top10,x="Total Revenue",y= "Product Name")
```

```python
# Top 5 Regions based on revenue
top5_regions = df.groupby("Region")["Total Revenue"].sum().nlargest(5).reset_index()
print("Top 5 Regions by Total Revenue:")
display(top5_regions)
```

```python
# Average Unit Price by Product Category
avg_price_by_category = df.groupby("Product Category")["Unit Price"].mean().reset_index()
print("\nAverage Unit Price by Product Category:")
display(avg_price_by_category)
```

```python
# Total Units Sold by Product Category
total_units_sold_by_category = df.groupby("Product Category")["Units Sold"].sum().reset_index()
print("\nTotal Units Sold by Product Category:")
display(total_units_sold_by_category)
```
